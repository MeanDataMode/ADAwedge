```python
# Function to remove quotes
def remove_quotes(lists):
    cleaned = []
    for item in lists:
        if item[0:1] == '"' and item[-1:] == '"':
            ab = item[:0] + item[(0+1):]
            ab = ab[:-1]
            cleaned.append(ab)
        elif item[0:1] == "'" and item[-1:]:
            ba = item[:0] + item[(0+1):]
            ba = ba[:-1]
            cleaned.append(ba)
        else:
            cleaned.append(item)
    return cleaned
```
